## Convert Pressure to Cp

```python
import pandas as pd
import numpy as np
```

```python
csv_pressure = "CAARC_example_CAARC_015/000/probes/hist_series/body_Pressure_body/bodies.body_Pressure_body.pressure.csv"
csv_atm = "CAARC_example_CAARC_015/000/probes/hist_series/point_Pressure_probe/points.point_Pressure_probe.pressure.csv"
Uh_ms = 70
rho_inf = 1.225

csv_cp_out = "CAARC_example_CAARC_015/000/probes/hist_series/body_Pressure_body/bodies.body_Pressure_body.cp.csv"
```

```python
df_pressure = pd.read_csv(csv_pressure)
df_pressure
```

```python
df_atm = pd.read_csv(csv_atm)
df_atm
```

```python
# Add atm column
df_pressure["atm"] = df_atm["0"]
df_pressure
```

```python
column_pts = [c for c in df_pressure.columns if c.isnumeric()]
len(column_pts)
```

```python
df_pressure[column_pts] = (df_pressure[column_pts].to_numpy().T - df_pressure["atm"].to_numpy()).T / (0.5 * rho_inf * (Uh_ms ** 2))
df_pressure[column_pts]
```

```python
df_pressure.to_csv(csv_cp_out, index=None)
```
