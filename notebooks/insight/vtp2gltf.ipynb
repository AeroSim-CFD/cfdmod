```python
import numpy as np
import pyvista as pv
import pathlib

mesh_p = pathlib.Path(
    "/home/ubuntu/Documentos/Repositories/cfdmod/output/pressure/cp/default/cp.stats.vtp"
)
mesh = pv.read(mesh_p)

pl = pv.Plotter()
_ = pl.add_mesh(
    mesh,
    cmap="viridis",
    scalars="min",
    smooth_shading=True,
    show_scalar_bar=False,
)
pl.export_gltf(mesh_p.parent / "v2.gltf")
pl.show()
```
